```python
import pandas as pd
```

```python
headers = pd.read_csv('headers_hourly02.txt', sep = ' ', header = 0)
```

```python
headers = headers.T
headers = headers[:-1]
headers_list = headers.iloc[:,0]
```

```python
headers_list
```

```python
df = pd.DataFrame()
for year in range(2004, 2024):
    tmp = pd.read_csv('./hourly02/'+ str(year) + '/CRNH0203-' + str(year) + '-NY_Ithaca_13_E.txt', delimiter= '\s+', header = None, na_values = -9999)
    df = pd.concat([df, tmp], axis = 0)
```

```python
df
```

```python
df.columns = headers_list
```
